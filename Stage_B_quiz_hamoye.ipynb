```
import pandas as pd
import matplotlib.pyplot as plt  #to plot data
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
import math

%matplotlib inline
```

```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
```

```
energy_df = pd.read_csv(url)
energy_df.head(5)
```

```
energy_df = energy_df.drop(['date', 'lights'], axis =1 )
energy_df
```

```
energy_df.columns = ['Appliances','T1',	'RH_1',	'T2',	'RH_2',	'T3',	'RH_3',	'T4',	'RH_4',	'T5',	'RH_5',	'T6',	'RH_6',	'T7',	'RH_7',	'T8',	'RH_8',	'T9',	'RH_9',	'T_out',	'Press_mm_hg',	'RH_out',	'Windspeed',	'Visibility',	'Tdewpoint',	'rv1',	'rv2']
```

```
scaler = MinMaxScaler()
energy_df = pd.DataFrame(scaler.fit_transform(energy_df), columns =  energy_df.columns)
energy_df
```

```
X = energy_df[['T2']]
Y = energy_df[['T6']]

# Split X and y into X_
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=42)

# create a Linear Regression model object
reg = LinearRegression()

# pass through the X_train & y_train data set
reg.fit(X_train, Y_train)
```

```
Y_predict = reg.predict(X_test)
```

```
reg_r2 = r2_score(Y_test, Y_predict)
round(reg_r2, 2)
```

```
energy_df_after = energy_df.drop(['Appliances'], axis = 1)

x = energy_df_after
y = energy_df['Appliances']

# Split X and y into X_
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)

# create a Linear Regression model object
reg = LinearRegression()

# pass through the X_train & y_train data set
reg.fit(x_train, y_train)
```

```
y_predict = reg.predict(x_test)
```

```
def get_weights_df(reg, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(reg.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

linear_model_weights = get_weights_df(reg, x_train, 'Linear_Model_Weight')
print(linear_model_weights)
```

```
model_mae = mean_absolute_error(y_test, y_predict)
print("MAE {:.2}".format(model_mae))
```

```
mse = mean_squared_error(y_test, y_predict)
round (mse, 3)
```

```
rmse =  math.sqrt(mse)
round(rmse, 3)
```

```
reg_r2 = r2_score(y_test, y_predict)
round (reg_r2, 2)
```

```
rss = np.sum(np.square(y_test - y_predict))
round(rss, 2)
```

```
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit (x_train, y_train)
```

```
predicted_value = lasso_reg.predict(x_test)
```

```
l_mse = mean_squared_error(y_test, predicted_value)
round(l_mse, 3)
```

```
rmse =  math.sqrt(l_mse)
round(rmse, 3)
```
